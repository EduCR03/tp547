```python
import scipy
from scipy.fftpack import fft
import numpy as np
from scipy import pi
import matplotlib.pyplot as plt
t = np.linspace(0,120,4000)


signal = 10*np.sin(2*pi*2.0*t) + 5*np.sin(2*pi*8.0*t) + 2*np.random.random(len(t))

FFT = abs(scipy.fftpack.fft(signal))
freqs = scipy.fftpack.fftfreq(signal.size, t[1]-t[0])

plt.subplot(211)
plt.plot(t, signal)
plt.subplot(212)
plt.plot(freqs,20*np.log10(FFT),'x')
plt.show()
```
