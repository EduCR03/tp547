```python
import numpy as np
import matplotlib.pyplot as plt

x=1
x1=np.array([x])
n=48
a=5
c=1
m=pow(2,6)-2
# m = 13
for i in range(n):
    x=(a*x+c)%m
    x1=np.append(x1,x)
print(x1)
ind=np.arange(n+1)
plt.bar(ind, x1)
plt.show()
```
