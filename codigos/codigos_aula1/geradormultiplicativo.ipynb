```python
import numpy as np
import matplotlib.pyplot as plt
# Gerador linear misto sem o c
x=5
x1=np.array([x])
n=60
a=3
m=16
for i in range(n):
    x=(a*x)%m
    x1=np.append(x1,x)
print(x1)
ind=np.arange(n+1)
plt.bar(ind, x1)
plt.xlabel('amostra')
plt.ylabel('X(i)')
plt.show()
```
