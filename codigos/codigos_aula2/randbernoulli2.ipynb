```python
import numpy as np
import matplotlib.pyplot as plt
p=0.5
av=np.array([])
x=np.random.uniform(0,1,1000000) # Gerador proprio do numpy
for ix in x:
    if ix >= p:
        a1=1
        av = np.append(av, a1)
    else:
        a1=0
        av=np.append(av,a1)
print(x)
print(av)
plt.hist(x,bins=20)
plt.show()
```
