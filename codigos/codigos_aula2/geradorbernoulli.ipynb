```python
import numpy as np
import matplotlib.pyplot as plt

x=1
x1=np.array([x])
av=np.array([]) # Gerar das 10000 os resultados da bernouli, distribuicao de bernouli =
n=10000
a=pow(7,5)
m=pow(2,31)-1
for i in range(n):
    x=(a*x)%m
    x1=np.append(x1,x) 
# ^ ------------- Gerando numeros aleatorios ------------ ^
x1=x1/m
for ix in x1:
    if ix >= 0.9:
        a1=1
        av = np.append(av, a1)
    else:
        a1=0
        av=np.append(av,a1)
print(x1)
print(np.mean(x1))
print(av)
ind=np.arange(n+1)
plt.bar(ind, av)
plt.show()
```
