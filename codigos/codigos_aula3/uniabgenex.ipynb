```python
import numpy as np

A=0
B=40
N=100000
value=21
count=0
x=np.random.uniform(0,1,N)

Xab=A+(B-A)*x
print(Xab)

for uniabvalue in Xab:
    if uniabvalue<=value:
        count=count+1
prob=count/N
print("a probabilidade e",prob)

Probana=(value-A)/(B-A)
print("a probabilidade analítica e",Probana)

import matplotlib.pyplot as plt
X = np.arange(A, B, 0.01)
fx = (1/(B-A)) * np.ones(np.size(X))
plt.hist(Xab, bins=50, density=True)
plt.plot(X, fx, 'r-', linewidth=2)
plt.title('Histograma dos valores gerados')
plt.xlabel('Valor')
plt.ylabel('Densidade')
plt.show()
```
