```python
import numpy as np
import matplotlib.pyplot as plt

d=3
b=4
q =d/b
x = np.array([-1.2,-0.2, -0.5, 0.4, 0.89, 1.3])

# Mid-tread
y = q * np.round(x/q)
erro_mid_tread = y - x

# Mid-riser
y2 = q * np.floor(x/q)+q/2
erro_mid_riser = y2 - x

print('Entrada original:')
print(x)

print('\nSaida quantizada mid-tread:')
print(y)
print('Erro de quantizacao mid-tread:')
print(erro_mid_tread)
print('Erro medio quadratico mid-tread:')
print(np.mean(erro_mid_tread**2))

print('\nSaida quantizada mid-riser:')
print(y2)
print('Erro de quantizacao mid-riser:')
print(erro_mid_riser)
print('Erro medio quadratico mid-riser:')
print(np.mean(erro_mid_riser**2))
```
