```python
import numpy as np
import matplotlib.pyplot as plt

N=10000
R=10
X=np.array([])
Y=np.array([])
for i in range(1,N):
    angulo = np.random.uniform(0,1)*2*np.pi
    raio= R*np.sqrt(np.random.uniform(0,1))
    x = raio * np.cos(angulo)
    y = raio * np.sin(angulo)
    X = np.append(X, x)
    Y = np.append(Y, y)

plt.plot(X,Y,'o')
plt.show()
```
